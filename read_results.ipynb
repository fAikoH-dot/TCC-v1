```python
import pandas as pd
```

```python
soja = pd.read_csv("Soja/Soja_Resultados.csv", sep=",")
soja["Cultivo"] = "Soja"
milho = pd.read_csv("Milho/Milho_Resultados.csv", sep=",")
milho["Cultivo"] = "Milho"
trigo = pd.read_csv("Trigo/Trigo_Resultados.csv", sep=",")
trigo["Cultivo"] = "Trigo"
algodao = pd.read_csv("Algodâo/Algodao_Resultados.csv", sep=",")
algodao["Cultivo"] = "Algodao"

soja.rename(columns={"ds":"Data", "y":"Estadual", "yhat":"Previsão"}, inplace=True)
milho.rename(columns={"ds":"Data", "y":"Estadual", "yhat":"Previsão"}, inplace=True)
trigo.rename(columns={"ds":"Data", "y":"Estadual", "yhat":"Previsão"}, inplace=True)
algodao.rename(columns={"ds":"Data", "y":"Estadual", "yhat":"Previsão"}, inplace=True)

soja["model"] = "PROPHET"
milho["model"] = "PROPHET"
trigo["model"] = "PROPHET"
algodao["model"] = "PROPHET"

soja2 = pd.read_csv("Soja/arima/Soja_Resultados_ARIMA.csv", sep=",")
milho2 = pd.read_csv("Milho/arima/Milho_Resultados_ARIMA.csv", sep=",")
trigo2 = pd.read_csv("Trigo/arima/Trigo_Resultados_ARIMA.csv", sep=",")
algodao2 = pd.read_csv("Algodâo/arima/Algodao_Resultados_ARIMA.csv", sep=",")

datas2 = pd.date_range(start='2004-03-01', end='2024-08-01', freq="MS")
soja2["Data"] = datas2
milho2["Data"] = datas2
trigo2["Data"] = datas2
algodao2["Data"] = datas2

soja2.rename(columns={"predict_Estadual":"Previsão"}, inplace=True)
milho2.rename(columns={"predict_Estadual":"Previsão"}, inplace=True)
trigo2.rename(columns={"predict_Estadual":"Previsão"}, inplace=True)
algodao2.rename(columns={"predict_Estadual":"Previsão"}, inplace=True)

soja2 = soja2[["Data", "Estadual", "País", "USD", "Previsão", "Cultivo", "model"]]
milho2 = milho2[["Data", "Estadual", "País", "USD", "Previsão", "Cultivo", "model"]]
trigo2 = trigo2[["Data", "Estadual", "País", "USD", "Previsão", "Cultivo", "model"]]
algodao2 = algodao2[["Data", "Estadual", "País", "USD", "Previsão", "Cultivo", "model"]]

soja2.fillna(0, inplace=True)
milho2.fillna(0, inplace=True)
trigo2.fillna(0, inplace=True)
algodao2.fillna(0, inplace=True)

soja2 = soja2[:197]
milho2 = milho2[:197]
trigo2 = trigo2[:197]
algodao2 = trigo2[:197]

datas = pd.date_range(start='2020-08-01', end='2026-09-01', freq="MS")
soja["Data"] = datas
milho["Data"] = datas
trigo["Data"] = datas

datas_alg = pd.date_range(start='2020-10-01', end='2026-09-01', freq="MS")
algodao["Data"] = datas_alg

soja = pd.concat([soja2,soja], ignore_index=True).reset_index(drop=True)
soja["model"] = "PROPHET"
milho = pd.concat([milho2,milho], ignore_index=True).reset_index(drop=True)
milho["model"] = "PROPHET"
trigo = pd.concat([trigo2,trigo], ignore_index=True).reset_index(drop=True)
trigo["model"] = "PROPHET"
algodao = pd.concat([algodao2,algodao], ignore_index=True).reset_index(drop=True)
algodao["model"] = "PROPHET"
```

```python
resultados_prophet = pd.concat([soja, milho, trigo, algodao])
# resultados_prophet["model"] = "PROPHET"
# resultados_prophet.rename(columns={"ds":"Data", "y":"Estadual", "yhat":"Previsão"}, inplace=True)
# resultados['ds'] = pd.to_datetime(resultados["ds"])
resultados_prophet['Estadual'] = resultados_prophet['Estadual'].map('{:,.2f}'.format)
resultados_prophet['País'] = resultados_prophet['País'].map('{:,.2f}'.format)
resultados_prophet['USD'] = resultados_prophet['USD'].map('{:,.2f}'.format)
resultados_prophet['Previsão'] = resultados_prophet['Previsão'].map('{:,.2f}'.format)
resultados_prophet
```

```python
resultados_prophet.to_csv("resultados_prophet.csv", sep=",", index=False)
```

```python
soja = pd.read_csv("Soja/arima/Soja_Resultados_ARIMA.csv", sep=",")
milho = pd.read_csv("Milho/arima/Milho_Resultados_ARIMA.csv", sep=",")
trigo = pd.read_csv("Trigo/arima/Trigo_Resultados_ARIMA.csv", sep=",")
algodao = pd.read_csv("Algodâo/arima/Algodao_Resultados_ARIMA.csv", sep=",")

datas = pd.date_range(start='2004-03-01', end='2024-08-01', freq="MS")
soja["Data"] = datas
milho["Data"] = datas
trigo["Data"] = datas
algodao["Data"] = datas
```
